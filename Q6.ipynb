```python
#name: Aijaz Hussain , Roll no. : 211103058, Branch: Computer Science
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('student_data.csv')

# Plot score distributions
subjects = ['math_score', 'english_score', 'history_score']
for subject in subjects:
    plt.figure(figsize=(8, 4))
    plt.hist(data[subject], bins=10, edgecolor='black')
    plt.title(f'{subject.capitalize()} Distribution')
    plt.xlabel('Scores')
    plt.ylabel('Number of Students')
    plt.grid(True)
    plt.show()

# Optional: Calculate and display basic statistics
for subject in subjects:
    print(f'\n{subject.capitalize()} Statistics:')
    print(data[subject].describe())
```
