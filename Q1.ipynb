```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from CSV file
df = pd.read_csv('student_data.csv')

# Inspect the data
print("Data Overview:")
print(df.head())
print("\nData Information:")
print(df.info())
print("\nDescriptive Statistics:")
print(df.describe())
print("\nMissing Values:")
print(df.isnull().sum())

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a figure with subplots
fig, axs = plt.subplots(2, 5, figsize=(20, 10))

# List of subjects to plot
subjects = ['math_score', 'science_score', 'english_score', 'history_score', 'attendance_rate']

# Plot histograms
for i, subject in enumerate(subjects):
    ax = axs[0, i]
    sns.histplot(df[subject], bins=20, kde=False, ax=ax)
    ax.set_title(f'Histogram of {subject}')
    ax.set_xlabel(subject)
    ax.set_ylabel('Frequency')

# Plot box plots
for i, subject in enumerate(subjects):
    ax = axs[1, i]
    sns.boxplot(x=df[subject], ax=ax)
    ax.set_title(f'Box Plot of {subject}')
    ax.set_xlabel(subject)

# Adjust layout
plt.tight_layout()
plt.show()
```
