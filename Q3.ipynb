```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from CSV file
data = pd.read_csv('student_data.csv')

# Calculate the average score across all subjects for each student
data['average_score'] = data[['math_score', 'english_score', 'history_score']].mean(axis=1)

# Sort the students based on the average score in descending order
top_students = data.sort_values(by='average_score', ascending=False)

# Select the top 10 students
top_10_students = top_students.head(10)

# Plot the bar chart for top 10 students' average scores
plt.figure(figsize=(12, 8))
plt.barh(top_10_students['name'], top_10_students['average_score'], color='skyblue')
plt.xlabel('Average Score')
plt.title('Top 10 Students by Average Score')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest scores at the top
plt.grid(axis='x')

# Show the plot
plt.show()

# Plot individual subject scores for top 10 students
fig, ax = plt.subplots(figsize=(14, 10))

# Plot Math scores
bar_width = 0.2
index = range(len(top_10_students))

bar1 = ax.barh(index, top_10_students['math_score'], bar_width, label='Math Score', color='lightcoral')
bar2 = ax.barh([i + bar_width for i in index], top_10_students['english_score'], bar_width, label='English Score', color='lightgreen')
bar3 = ax.barh([i + 2 * bar_width for i in index], top_10_students['history_score'], bar_width, label='History Score', color='lightblue')

# Labeling
ax.set_xlabel('Scores')
ax.set_title('Top 10 Students’ Scores by Subject')
ax.set_yticks([i + bar_width for i in index])
ax.set_yticklabels(top_10_students['name'])
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
