```python
#name: Aijaz Hussain , Roll no. : 211103058, Branch: Computer Science
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('student_data.csv')

# Display the first few rows of the dataframe to understand its structure
df.head()

# Calculate average scores by gender
average_scores_by_gender = df.groupby('gender').agg({
    'math_score': 'mean',
    'english_score': 'mean',
    'history_score': 'mean'
}).reset_index()

# Print the average scores by gender
print("Average Scores by Gender:")
print(average_scores_by_gender)

# Plot average scores by gender
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# Math Score Plot
ax[0].bar(average_scores_by_gender['gender'], average_scores_by_gender['math_score'], color=['blue', 'pink'])
ax[0].set_title('Average Math Scores by Gender')
ax[0].set_xlabel('Gender')
ax[0].set_ylabel('Average Math Score')

# English Score Plot
ax[1].bar(average_scores_by_gender['gender'], average_scores_by_gender['english_score'], color=['blue', 'pink'])
ax[1].set_title('Average English Scores by Gender')
ax[1].set_xlabel('Gender')
ax[1].set_ylabel('Average English Score')

# History Score Plot
ax[2].bar(average_scores_by_gender['gender'], average_scores_by_gender['history_score'], color=['blue', 'pink'])
ax[2].set_title('Average History Scores by Gender')
ax[2].set_xlabel('Gender')
ax[2].set_ylabel('Average History Score')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('student_data.csv')
gender_scores = df.groupby('gender')[['math_score', 'english_score', 'history_score']].mean()
gender_scores.plot(kind='bar', figsize=(10, 6))
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.legend(['Math', 'English', 'History'])
plt.show()
```
